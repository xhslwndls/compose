```python
artist_list=['my chemical romance','bullet for my valentine','avenged sevenfold',
             'slipknot','linkin park','paramore',
             'halestorm','sum 41','lamb of god',
             'trivium','killswitch engage','dragonforce',
             'black veil brides','three days grace','system of a down']
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib.request

def song_match(artist_input):
    song_matcher=[]
    driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver')
    driver.implicitly_wait(3)
    artist_inp=artist_input
    artist = artist_inp.replace(" ", "%20")
    driver.maximize_window()
    driver.get('https://vibe.naver.com/search/tracks?query='+artist)
    driver.find_element_by_xpath("//*[@id='app']/div[2]/div/div/a[2]").click()
    time.sleep(1)
    count=0
    while True:
        print(count)
        if count>5:
            break
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(0.3)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        try:
            time.sleep(0.5)
            driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[3]/a").click()               
        except:    
            try:
                driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[5]/a").click() 
            except:
                try:
                    driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[6]/a").click() 
                    #평상시 6, LoG 때 4
                except:           
                    try:
                        driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[10]/a").click()
                    except:
                        try:
                            driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[4]/a").click()
                        except:
                            try:
                                driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[8]/a").click()
                            except:
                                count+=1
    num_song_elem=driver.find_element_by_xpath("//*[@id='content']/div/div[2]/div[2]/div[1]/div[2]/div/span")
    num_song=int(re.findall('[0-9]*',num_song_elem.text.replace(',',''))[0])
    for i in range(1,num_song+1):
        #print(1)
        try:
            artist_name_find="//*[@id='content']/div/div[2]/div[2]/div[2]/div/table/tbody/tr[{ind}]/td[4]/span/span/span/a/span".format(ind=i)
            artist_name=driver.find_element_by_xpath(artist_name_find).text
        except:
            #print(2)
            continue
        #print(artist_name.lower(),artist_inp.lower())
        if artist_name.lower()==artist_inp.lower():
            #print(1)
            inf_find="//*[@id='content']/div/div[2]/div[2]/div[2]/div/table/tbody/tr[{ind}]/td[1]".format(ind=i)
            info_temp=driver.find_element_by_xpath(inf_find)
            info=re.findall('chk_track_(.*)',info_temp.find_elements_by_xpath(".//*")[0].get_attribute('id'))
            #print(info)
            song_name_find="//*[@id='content']/div/div[2]/div[2]/div[2]/div/table/tbody/tr[{ind}]/td[3]/div[1]/span/a".format(ind=i)
            song_name=driver.find_element_by_xpath(song_name_find).text
            print(song_name,info,artist_name)
            song_matcher.append([info[0],artist_name,song_name])
    #print(artist_input,"done")
    return song_matcher
```

```python
artist_song_match=[]
#012완료 3완료 4완료 5완료 6완료 7완료 8 9완료 10완료 11완료 12완료 13완료 14완료
for x in artist_list[8:9]:
    matched_song_list=song_match(x)
    artist_song_match.append(matched_song_list)
```
