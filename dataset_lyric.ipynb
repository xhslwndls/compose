```python
import re
import pickle
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
```

```python
artist_list=['my chemical romance','bullet for my valentine','avenged sevenfold',
             'slipknot','linkin park','paramore',
             'halestorm','sum 41','lamb of god',
             'trivium','killswitch engage','dragonforce',
             'black veil brides','three days grace','system of a down']
```

```python
def crawl_lyric(song_num):
    driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver')
    driver.implicitly_wait(3)
    driver.maximize_window()
    driver.get('https://vibe.naver.com/track/'+song_num)
    try:
        song_lyric=driver.find_element_by_xpath("//*[@id='content']/div[3]/p").text
        return song_lyric
    except:
        return 'No Lyrics'
    driver.close()
```

```python
with open('artist_song_match_012.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
song_matched_lyric=[]
#0 1 2 3 4완료 5완료 6완료 7완료 8완료 9완료 10완료 11완료 12완료 13완료 14완료 
for song_data in data[0]:
    song_data_temp=song_data
    song_num=song_data[0]
    try:
        song_lyric=crawl_lyric(song_num)
    except:
        continue
    if song_lyric!='No Lyrics':
        song_data_temp.append(song_lyric)
        song_matched_lyric.append(song_data_temp)
```
